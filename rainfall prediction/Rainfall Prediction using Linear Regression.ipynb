```python
#importing the library
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("austin_weather_final.csv")
data
```

```python
X=data.drop(['PrecipitationSumInches'],axis=1)
```

```python
Y=data["PrecipitationSumInches"]
```

```python
#reshaping it to 2d vector
Y=Y.values.reshape(-1,1)
```

```python
day_index=798
days=[i for i in range(Y.size)]
```

```python
#initializing the linear regression classifier
clf= LinearRegression()
#train the classifier
clf.fit(X,Y)
```

```python
#plotting a graph
print("The precipitation trend graph")
plt.scatter(days,Y,color='g')
plt.scatter(days[day_index],Y[day_index],color='r')
plt.title("Precipitation Level")
plt.xlabel("Days")
plt.ylabel("Preciptation in inches")
plt.show()
x_vis=X.filter(['TempAvgf','DewPointAvgF','HumidityAvgPercent','SeaLevelPressureAvgInches','VisiblityAvgMiles','WindAvgMPH'])
```

```python
print("The precipitation vs attributes trend graph")
for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days,x_vis[x_vis.columns.values[i][:100]],color='g')
    plt.scatter(days[day_index],x_vis[x_vis.columns.values[i]][day_index],color='r')
    plt.title(x_vis.columns.values[i])
plt.show()
```
